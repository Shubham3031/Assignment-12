```python
# 1.Ans.

When creating a custom exception in Python, it is recommended to use the built-in 
Exception class as the base class for your custom exception. 
This is because Exception is a general-purpose base class for all built-in, 
non-system-exiting exceptions in Python. Using Exception as the base class 
allows your custom exception to inherit all the useful attributes and methods of the Exception class.

By inheriting from Exception, your custom exception can be caught by any try-except 
block that catches Exception or any of its subclasses, without having to specifically 
catch your custom exception. This makes it easier to handle errors and exceptions in
your code, as you can catch multiple types of exceptions with a single except statement.

   Example :
        try:
    # some code that might raise MyException
except MyException as e:
    # handle MyException
except Exception as e:
    # handle any other exception
```

```python
# 4.Ans.

The LookupError class in Python is a built-in exception class 
that is raised when a key or index used to access a collection 
(such as a dictionary or a list) is not found. It is a base class
for specific lookup error exceptions such as KeyError and IndexError.

KeyError is a specific exception that is raised when a key is not found in a dictionary.
        For example:
        
        my_dict = {'a': 1, 'b': 2, 'c': 3}
print(my_dict['d'])

IndexError is a specific exception that is raised when an index used to access
a list or tuple is out of range. 
              For example:
        
        my_list = [1, 2, 3]
print(my_list[3])


```

```python
# 5.Ans.

ImportError is a built-in exception in Python that is raised when a module,
package, or name that is being imported cannot be found or loaded.
This can occur for a variety of reasons, such as a typo in the name of the module, 
a missing dependency, or an issue with the Python path or environment.

            import some_module
    
In Python 3.6 and later versions, a new more specific exception called
ModuleNotFoundError is introduced. ModuleNotFoundError is raised when a module
is not found, just like ImportError, but it provides more detailed information 
about the module that could not be found.

       import some_module_that_does_not_exist

```

```python
# 6.Ans.

Be specific with your except statements: Avoid using a bare except 
statement (except:) as it catches all exceptions including system errors
and keyboard interrupts which can hide serious issues. 
Instead, use specific exception types to catch only the exceptions that you anticipate.

Use finally for cleanup code: If you have cleanup code that needs to be 
run regardless of whether an exception occurs or not, put it in a finally block. 
This block will always execute, even if an exception is raised in the try block.

Use the with statement: If you are working with resources that need to be 
closed when you are done with them (like files or network connections), 
use the with statement. This will automatically handle closing the resource, 
even if an exception is raised.

Use built-in exceptions when possible: Python has many built-in exceptions that cover most common use cases. 
Whenever possible, use these instead of creating your own custom exceptions.
```
